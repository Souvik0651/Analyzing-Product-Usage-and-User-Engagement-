##**Analyzing Product Usage and User Engagement**

```python
# importing pandas library
import pandas as pd
```

```python
# importing excel files from drive
rec=pd.read_excel('/content/drive/MyDrive/Vantage Assignment/receiver.xlsx')
rec
```

```python
# dataset information
rec.info()
# changing data type 
rec['receiver_id']=rec['receiver_id'].astype(str)
rec['sender_id']=rec['sender_id'].astype(str)
```

```python
rec.info()
```

```python
# importing dataset sender
sen=pd.read_excel('/content/drive/MyDrive/Vantage Assignment/sender.xlsx')
sen['sender_id']=sen['sender_id'].astype(str)
```

```python
# importing dataset manager
man=pd.read_excel('/content/drive/MyDrive/Vantage Assignment/manager.xlsx')
man['manager_id']=man['manager_id'].astype(str)
```

```python
# merging recviver with manager dataset
df=pd.merge(rec,man,left_on='sender_id',right_on='manager_id')
```

```python
df
```

```python
len(sen.sender_id.unique())
```

```python
len(man)
```

```python
# merging sender with manger
df2=pd.merge(sen,man,left_on='sender_id',right_on='manager_id',how='right')
df2=df2.drop_duplicates()
df2.shape
```

Top Receiver

```python
#top receivers
rec1=rec.drop_duplicates() #removing duplicates
#group by receiver_id with points 
top_rec = rec1.groupby('receiver_id')['points'].sum()
```

```python
# resetting index
top_rec=top_rec.reset_index()
```

```python
# top 10 receiver
top_rec.sort_values(by=['points'],ascending=False).head(10)
```

Top awarding managers

```python
# filtering awards from feed type and groupby feedtype with manager
top_awards = df2[df2['feed type']=='award'].groupby('manager_id')['feed type'].count()
```

```python
# top 5 awarding manager
top_awards.reset_index().sort_values(by=['feed type'], ascending=False).head(5)
```

Monthly analysis of total points received

```python
#monthly analysis
rec1.info()
```

```python
# function made for finding month
def find_month(x):
  return x.month
```

```python
rec1['month']=rec1['date'].apply(find_month)
```

```python
rec1
```

```python
# finding unique values for month
rec1.month.unique()
```

```python
# groupby month with points
rec1.groupby(['month'])['points'].sum()
```

Avg. points received by the users per week

```python
# function for finding week
def find_week(x):
  return x.week
```

```python
rec1['week']=rec1['date'].apply(find_week)
```

```python
rec1
```

```python
# group by week with points to find weekly average
avq_week=rec1.groupby(['week'])['points'].mean()
```

```python
avq_week=avq_week.reset_index()
```

```python
# renaming column
avq_week=avq_week.rename(columns={'points':'avg_per week'})
```

```python
avq_week
```

Top countries on the basis of active users

```python
# groupby country with receiver_id
top_coun=rec1.groupby('country')['receiver_id'].count()
```

```python
top_coun.reset_index().sort_values(by=['receiver_id'],ascending=False)
```
